# Explore and Clean Flight Dataset

Data from: https://zenodo.org/record/4670228#.YIL5Y0UzaUk

```python
import pandas as pd
import numpy as np
```

## Load Data

```python
flights = pd.read_csv("flightlist_20190101_20190131.csv")
print(flights.shape)
flights.head(10)
```

## Get Relevant Columns

```python
flights.columns
```

```python
flights = flights[["day","origin", "destination"]]
flights = flights.dropna()
flights = flights.sample(1000, random_state = 10)
print(flights.shape)
flights.head()
```

```python
unique = np.unique(np.append(flights["origin"], flights["destination"]))
print(unique.shape)
unique
```

```python
flights.to_csv("flights_sample.csv", index = False)
```
